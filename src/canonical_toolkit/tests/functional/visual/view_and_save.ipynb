thumbnails from memory

```python
thumbnails = snapshot_thumbnail_converter(scale=0.5)
```

```python
def show_base64_image(base64_string: str, width: int | None = None):
    """Display a base64-encoded image in a Jupyter notebook."""
    from IPython.display import HTML, display
    width_style = f'width="{width}"' if width else ""
    html = f'<img src="{base64_string}" {width_style}/>'
    display(HTML(html))
```

```python
show_base64_image(thumbnails[0])
```
