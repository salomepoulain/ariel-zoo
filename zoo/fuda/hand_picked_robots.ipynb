```python
import canonical_toolkit as ctk
```

### BabyA

```python
baby_a = ctk.node_from_string('C[b(H2B)f(H2H2BH2B)r(HBHB[rl(H2B)])]')
```

```python
baby_a.shift_visual_rotation(1)
ctk.quick_view(baby_a.to_graph())
print(baby_a.canonicalize(limb_order=False).to_string())
print(baby_a.copy().canonicalize().to_string())
```

```python
baby_a_aligned = "C[f(H2B)b(H2H2BH2B)l(HBHB[rl(H2B)])]"
```

```python
all_robots = [
    baby_a_aligned,
    baby_b_aligned,
    blokky_aligned,
    garrix_aligned,
    gecko_aligned,
    insect_aligned,
    linkin_aligned,
    longleg_aligned,
    penguin_aligned,
    pentapod_aligned,
    queen_aligned,
    salamander_aligned,
    snake_aligned,
    spider_aligned,
    squarish_aligned,
    stingray_aligned,
    tinlicker_aligned,
    turtle_aligned,
    ww_aligned,
    zappa_aligned,
]

hand_picked = [
    spider_aligned,
    gecko_aligned,
    baby_a_aligned,
    baby_b_aligned,
    snake_aligned,
]

import pandas as pd

all_df = pd.DataFrame({"fuda_aligned" : all_robots, "canonical": [ctk.node_from_string(robot).canonicalize().to_string() for robot in all_robots]})
all_df.to_csv('all_fuda.csv')

hand_picked_df = pd.DataFrame({"fuda_aligned" : hand_picked, "canonical": [ctk.node_from_string(robot).canonicalize().to_string() for robot in hand_picked]})
hand_picked_df.to_csv('hand_picked.csv')
```

```python
ctk.snapshot_saver([ctk.node_from_string(robot).to_graph() for robot in all_robots], save_folder="__data__/snapshots/all")
ctk.snapshot_saver([ctk.node_from_string(robot).to_graph() for robot in hand_picked], save_folder="__data__/snapshots/hand_picked")
```

```python
from canonical_toolkit.morphology.visual.plots import SnapshotGridPlotter

hand_picked_plotter = SnapshotGridPlotter.from_snapshot_folder("__data__/snapshots/hand_picked", grid_shape=(1, 5), snapshots_per_subplot=1)
all_plotter = SnapshotGridPlotter.from_snapshot_folder("__data__/snapshots/all", grid_shape=(4, 5), snapshots_per_subplot=1)
```

```python
hand_picked_plotter.plot()
```
