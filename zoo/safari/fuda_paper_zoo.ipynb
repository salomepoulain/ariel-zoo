```python
import canonical_toolkit as ctk
```

### BabyA

```python
baby_a = ctk.node_from_string('C[b(H2B)f(H2H2BH2B)r(HBHB[rl(H2B)])]')
```

```python
baby_a.shift_visual_rotation(1)
ctk.quick_view(baby_a.to_graph())
print(baby_a.canonicalize(limb_order=False).to_string())
print(baby_a.copy().canonicalize().to_string())
```

```python
baby_a_aligned = "C[f(H2B)b(H2H2BH2B)l(HBHB[rl(H2B)])]"
```

```python
import pandas as pd

all_robots = {
    "baby_a": baby_a_aligned,
    "baby_b": baby_b_aligned,
    "blokky": blokky_aligned,
    "garrix": garrix_aligned,
    "gecko": gecko_aligned,
    "insect": insect_aligned,
    "linkin": linkin_aligned,
    "longleg": longleg_aligned,
    "penguin": penguin_aligned,
    "pentapod": pentapod_aligned,
    "queen": queen_aligned,
    "salamander": salamander_aligned,
    "snake": snake_aligned,
    "spider": spider_aligned,
    "squarish": squarish_aligned,
    "stingray": stingray_aligned,
    "tinlicker": tinlicker_aligned,
    "turtle": turtle_aligned,
    "ww": ww_aligned,
    "zappa": zappa_aligned,
}

hand_picked = {
    "spider": spider_aligned,
    "gecko": gecko_aligned,
    "baby_a": baby_a_aligned,
    "baby_b": baby_b_aligned,
    "snake": snake_aligned,
}

all_df = pd.DataFrame({
    "trivial_name": list(all_robots.keys()),
    "fuda_aligned": list(all_robots.values()),
    "canonical": [ctk.node_from_string(robot).canonicalize().to_string() for robot in all_robots.values()]
})
all_df.to_csv('data/fuda_zoo/selected_20.csv', index=False)

hand_picked_df = pd.DataFrame({
    "trivial_name": list(hand_picked.keys()),
    "fuda_aligned": list(hand_picked.values()),
    "canonical": [ctk.node_from_string(robot).canonicalize().to_string() for robot in hand_picked.values()]
})
hand_picked_df.to_csv('data/hand_picked_5.csv', index=False)
```

```python
ctk.snapshot_saver([ctk.node_from_string(robot).to_graph() for robot in all_robots.values()], save_folder="data/fuda_zoo/snapshots/selected_20")
ctk.snapshot_saver([ctk.node_from_string(robot).to_graph() for robot in hand_picked.values()], save_folder="data/fuda_zoo/snapshots/hand_picked_5")
```

```python
hand_picked_plotter = ctk.GridPlotter()
hand_picked_plotter.add_image_data(data_folder="data/fuda_zoo/snapshots/hand_picked_5", shape=(1, 5))
hand_picked_plotter.show()
```
