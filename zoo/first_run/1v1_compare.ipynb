```python
from ariel_experiments.gui_vis.visualize_tree import visualize_tree_from_graph
from ariel_experiments.utils.initialize import generate_random_individual
from rich.console import Console
from ariel_experiments.characterize.canonical.core.toolkit import CanonicalToolKit as ctk
from ariel_experiments.gui_vis.view_mujoco import view
from PIL import Image
from IPython.display import display

console = Console()
```

generate and show subtrees robot 1

```python
random_robot1 = generate_random_individual()
view(random_robot1, with_viewer=True)
```

```python
view(random_robot1, remove_background=True, return_img=True, tilted=True)
graphs_robot1 = ctk.collect_neighbours(ctk.from_graph(random_robot1), do_radius_prefix=False, output_type=ctk.OutputType.GRAPH)
for graph in list(graphs_robot1.values())[2]:
    img = view(graph, tilted=True, remove_background=True, return_img=True)
    display(img)
```

```python
visualize_tree_from_graph(random_robot1)
for graph in list(graphs_robot1.values())[2]:
    visualize_tree_from_graph(graph)
```
